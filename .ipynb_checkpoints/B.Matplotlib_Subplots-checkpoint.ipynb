<h3>B Matplotlib Subplots</h3>

<p>Illustrative examples of 3 methods of creating subplots</p>

```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt

fib = (1, 8, 5, 3, 2,21,13)
x = np.linspace(1,len(fib),num=len(fib))
```

<p><font size="3" color="red">1. Using Subplots</font></p>

<p> subplot(122) = 1x2 grid, 2nd subplot<p>

```python
plt.figure(1,figsize=(12,4))  

plt.subplot(121)   
plt.plot(fib)
plt.title("Title A")

plt.subplot(122) 
plt.bar(x,fib)
plt.ylabel("Y Label: B")
```

<p><font size="3" color="red">2. Creating a figure object with two axes</font></p>

```python
fig, ((ax1,ax2)) = plt.subplots(1,2,figsize=(12, 4))

ax1.plot(fib)
ax1.set_title("Title A")
ax2.bar(x,fib)
ax2.set_ylabel("Y Label: B")
```

<p><font size="3" color="red">3. Using ravel(). Can pass conditions to ravel to control plots. </font></p>

```python
fig, axs = plt.subplots(1,2, figsize=(12, 4))

axs = axs.ravel()
for i in range(2):
    if i==0:
        axs[i].plot(fib)
    if i==1:
        axs[i].bar(x,fib)
    axs[i].set_title(('Title '+str(i)))
   
```
